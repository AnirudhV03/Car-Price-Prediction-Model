### DATA PREPROCESSING

```python
# FINDING THE AGE OF THE CAR USING DATETIME FUNCTION

age = date_time.year - car_data['Year']
car_data['Age'] = age
car_data.head()
```

### OUTLIER REMOVAL

```python
import seaborn as sns
```

```python
sns.boxplot(car_data['Selling_Price'])
```

```python
# From the above chart, we can see that there are 2 outliers. To display these data points, we will display the Selling_Price column in descending order.

sorted(car_data['Selling_Price'],reverse=True)
```

```python
# Displaying the 2 outlier rows
car_data[(car_data['Selling_Price']>=33.0) & (car_data['Selling_Price']<=35.0)]

# Updating car_data by skipping the 2 outlier rows. This is done by using the tilda sign. Hence, we have now removed the outlier rows from the dataset.
car_data = car_data[~(car_data['Selling_Price']>=33.0) & (car_data['Selling_Price']<=35.0)]
```

```python
car_data.shape
```

### ENOCDING THE CATEGORICAL COLUMNS

```python
#ML MODELS RECOGNIZE NUMERICAL DATA. HENCE WE NEED TO CONVERT THE CATEGORICAL DATA INTO NUMERICAL DATA.

car_data.head(1)
```

```python
# FINDING THE UNIQUE VALUES IN THE COLUMN Fuel_Type

car_data['Fuel_Type'].unique()
```

```python
# CONVERTING THE UNIQUE VALUES INTO NUMERICAL VALUES

#car_data['Fuel_Type'] = car_data['Fuel_Type'].map({'Petrol':0,'Diesel':1,'CNG':2},na_action='ignore')

# Creating user defined function to change the unique values to numericals
def my_function(x):
  if x == 'Petrol':
    return(0)
  if x == 'Diesel':
    return(1)
  if x == 'CNG':
    return(2)

# Mapping the unique values to numericals using the user defined function
car_data['Fuel_Type'] = car_data['Fuel_Type'].map(my_function)

car_data['Fuel_Type'].unique()
```

```python
car_data.head()
```

```python
car_data['Selling_type'].unique()
```

```python
# SIMILARLY, CONVERTING THE Selling_type COLUMN TO NUMERICAL DATA TYPE

def seller_type(x):
  if x == 'Dealer':
    return(0)
  if x == 'Individual':
    return(1)

# Mapping the unique values to numericals using the user defined function
car_data['Selling_type'] = car_data['Selling_type'].map(seller_type)

car_data['Selling_type'].unique()
```

```python
car_data.head()
```

```python
car_data['Transmission'].unique()
```

```python
# SAME PROCEDURE AS ABOVE FOR Transmission COLUMN

def transmission(x):
  if x == 'Manual':
    return(0)
  if x == 'Automatic':
    return(1)

# Mapping the unique values to numericals using the user defined function
car_data['Transmission'] = car_data['Transmission'].map(transmission)

car_data['Transmission'].unique()
```

### STORE FEATURE MATRIX IN X AND RESPONSE (TARGET) IN VECTOR Y

```python
x = car_data.drop(['Car_Name','Selling_Price'],axis=1) # x now contains the data series with the 2 columns excluded
y = car_data['Selling_Price'] # y now contains the Selling_Price column
```

```python
y
```

### SPLITTING THE DATASET INTO TRAINING SET AND TEST SET

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print(x_train.shape,y_train.shape,x_test.shape,y_test.shape)
```

### IMPORT THE MODELS

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from xgboost import XGBRegressor
```

### MODEL TRAINING

```python
lr = LinearRegression()
lr.fit(x_train,y_train)

rf = RandomForestRegressor()
rf.fit(x_train,y_train)

gbr = GradientBoostingRegressor()
gbr.fit(x_train,y_train)

xgb = XGBRegressor()
xgb.fit(x_train,y_train)
```

### PREDICTION ON TEST DATA

```python
y_pred1 = lr.predict(x_test)
y_pred2 = rf.predict(x_test)
y_pred3 = gbr.predict(x_test)
y_pred4 = xgb.predict(x_test)
```

### EVALUATING THE ALGORITHM

```python
from sklearn import metrics
```

```python
score1 = metrics.r2_score(y_test,y_pred1)
score2 = metrics.r2_score(y_test,y_pred2)
score3 = metrics.r2_score(y_test,y_pred3)
score4 = metrics.r2_score(y_test,y_pred4)
```

```python
print(score1*100,score2*100,score3*100,score4*100)
```

```python
final_scores = pd.DataFrame({'Models':['LR','RF','GBR','XGR'],
                             'R2_SCORE':[score1*100,score2*100,score3*100,score4*100]})

final_scores
```

```python
sns.barplot(x=final_scores['Models'],y=final_scores['R2_SCORE'])
```
